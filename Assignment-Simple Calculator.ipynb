```python
from tkinter import *
window=Tk()
```

```python
window.title("Calculator")
window.geometry('250x300')

expression = ""

def press(num):
    
    global expression
    expression=expression + str(num)
    equation.set(expression)
    
def equalpress():
    try:
        global expression
        total=str(eval(expression))
        equation.set(total)
    
        expression=""
    except:
        equation.set("Error")
        expression=""
        
def clear():
    global expression
    expression=""
    equation.set("")
    

equation=StringVar()

expression_field=Entry(window,textvariable=equation)
expression_field.grid(columnspan=4)
    

button1=Button(window,text='1',command=lambda: press(1),height=1,width=6)
button1.grid(column=0, row=1 )

button2=Button(window,text='2',command=lambda: press(2),height=1,width=6)
button2.grid(column=1, row=1 )

button3=Button(window,text='3',command=lambda: press(3),height=1,width=6)
button3.grid(column=2, row=1 )

button4=Button(window,text='4',command=lambda: press(4),height=1,width=6)
button4.grid(column=0, row=2 )

button5=Button(window,text='5',command=lambda: press(5),height=1,width=6)
button5.grid(column=1, row=2 )

button6=Button(window,text='6',command=lambda: press(6),height=1,width=6)
button6.grid(column=2, row=2 )

button7=Button(window,text='7',command=lambda: press(7),height=1,width=6)
button7.grid(column=0, row=3 )

button8=Button(window,text='8',command=lambda: press(8),height=1,width=6)
button8.grid(column=1, row=3 )

button9=Button(window,text='9',command=lambda: press(9),height=1,width=6)
button9.grid(column=2, row=3 )

button0=Button(window,text='0',command=lambda: press(0),height=1,width=6)
button0.grid(column=1, row=4 )

plus=Button(window,text='+',command=lambda: press("+"),height=1,width=6)
plus.grid(column=3, row=1)

minus=Button(window,text="-",command=lambda: press("-"),height=1,width=6)
minus.grid(column=3, row=2)

multiply=Button(window,text="*",command=lambda: press("*"),height=1,width=6)
multiply.grid(column=3,row=3)

divide=Button(window,text="/",command=lambda: press("/"),height=1,width=6)
divide.grid(column=3,row=4)

equal=Button(window,text="=",command=equalpress, height=1,width=6)
equal.grid(column=2,row=4)

decimal=Button(window,text=".",command=lambda: press("."), height=1,width=6)
decimal.grid(column=0,row=4)

clear=Button(window,text="Clear",command=clear, height=1,width=6)
clear.grid(column=1,row=5)

window.mainloop()
```
